```python
#importanto Biblioteca no Python
import pandas as pd

combustiveis_df = pd.read_excel("ca-2021-02.xlsx")
```

```python
#describe() mostra estatisticas 'mais básicas' que a gente precisa
print(combustiveis_df.describe())
```

```python
#Quais sao as colunas e que tipo de dados cada um tem
print(combustiveis_df.info())
```

```python
#filtrar apenas por uma coluna
display(combustiveis_df['Revenda'])
```

```python
#Aqui criamos um novo dataframe apenas com as colunas que eu quero...
ca_df = combustiveis_df[['Revenda','Municipio','Produto','Valor de Venda']]
display(ca_df)
```

```python
#Exibe a 4a linha
display(ca_df.loc[3])
```

```python
#Exibir da 10a. linha até a 20a.linha
display(ca_df.loc[9:19])
```

```python
#Criar uma Dataframe gas_df contendo apenas as 4 colunas(Revenda, Municipio, Produto, Valor de Venda) 
# somente com combustivel sendo Gasolina.
gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df)

display(gas_df['Valor de Venda'].max())
```

```python
#Exibir qual é o posto e municipio da gasolina mais cara do pais?
display(gas_df[['Revenda','Municipio','Valor de Venda']].max())
```

```python
#Dataframe.[] com multiplas condiçoes para filtragem
#Quais são os preços de ETANOL na minha cidade(INDAIATUBA)? ordenado do menor preço para o maior
etanol_indaiatuba_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio']=='INDAIATUBA')]
display(etanol_indaiatuba_df.sort_values(by = 'Valor de Venda'))
```

```python
#Qual a media dos preços dos combustiveis GASOLINA e GASOLINA ADITIVADA do bairro BELA VISTA em São Paulo
combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & (combustiveis_df['Municipio'] == 'SAO PAULO')]
display(combustiveis_df)
```
